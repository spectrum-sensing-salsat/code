```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import signal
from scipy import stats
from scipy import linalg

from mpl_toolkits import mplot3d as mpl3d
```

```python
mpl.use("pgf")
mpl.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

zoom = 0.60
line = 0.75
```

```python
x = np.linspace(-5, 5, 1000)
plt.xlim(-3.9, 3.9)
plt.ylim(-0.05, 0.45)
plt.plot(x, stats.norm.pdf(x), 'k-', linewidth=line, antialiased=True)
plt.xlabel(r'$x$')
plt.ylabel(r'$p(x)$')
plt.locator_params(axis='y', nbins=6)
plt.locator_params(axis='x', nbins=10)
plt.grid(linewidth=0.3)
fig = mpl.pyplot.gcf()
w, h = fig.get_size_inches()
fig.set_size_inches(w*zoom, h*zoom)

plt.savefig('out/norm_pdf.pgf')
```

```python
x = np.linspace(-5, 5, 1000)
plt.xlim(-3.9, 3.9)
plt.ylim(-0.05, 1.05)
plt.plot(x, stats.norm.cdf(x), 'k-', linewidth=line, antialiased=True)
plt.xlabel(r'$x$')
plt.ylabel(r'$\Phi(x)$')
plt.locator_params(axis='y', nbins=6)
plt.locator_params(axis='x', nbins=10)
plt.grid(linewidth=0.3)
fig = mpl.pyplot.gcf()
w, h = fig.get_size_inches()
fig.set_size_inches(w*zoom, h*zoom)

plt.savefig('out/norm_cdf.pgf')
```

```python
x = np.linspace(-5, 5, 1000)
plt.xlim(-3.9, 3.9)
plt.ylim(-0.05, 1.05)
plt.plot(x, stats.norm.sf(x), 'k-', linewidth=line, antialiased=True)
plt.xlabel(r'$x$')
plt.ylabel(r'$Q(x)$')
plt.locator_params(axis='y', nbins=6)
plt.locator_params(axis='x', nbins=10)
plt.grid(linewidth=0.3)
fig = mpl.pyplot.gcf()
w, h = fig.get_size_inches()
fig.set_size_inches(w*zoom, h*zoom)

plt.savefig('out/norm_ccdf.pgf')
```

```python
x = np.linspace(0, 12, 1000)
plt.xlim(0, 10.5)
plt.ylim(0, 0.5)
plt.plot(x, stats.chi2.pdf(x, df=1), 'k-', linewidth=line,antialiased=True)
plt.plot(x, stats.chi2.pdf(x, df=2), 'k--', linewidth=line, antialiased=True)
plt.plot(x, stats.chi2.pdf(x, df=5), 'k-.', linewidth=line, antialiased=True)
plt.plot(x, stats.chi2.pdf(x, df=10), 'k:', linewidth=line, antialiased=True)
plt.legend(('$N=1$', '$N=2$', '$N=5$', '$N=10$'), loc=0)
plt.xlabel(r'$x$')
plt.ylabel(r'$\chi_N^2(x)$')
plt.locator_params(axis='y', nbins=6)
plt.locator_params(axis='x', nbins=10)
plt.grid(linewidth=0.3)
fig = mpl.pyplot.gcf()
w, h = fig.get_size_inches()
fig.set_size_inches(w*zoom, h*zoom)

plt.savefig('out/chi_pdf.pgf')
```

```python
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))
mu = np.array([0, 0])
cov = np.array([[1, 0.5],[0.5, 1]])
rv = stats.multivariate_normal(mu, cov)
Z = rv.pdf(pos)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.15, linewidth=0.3, color='black', edgecolors=(0,0,0,1))
ax.view_init(elev=20, azim=20)
ax.set_xticks([-4, -2, 0, 2, 4])
# ax.set_yticks([-4, -2, 0, 2, 4])
# ax.set_zticks([0.04, 0.08, 0.12])

ax.zaxis.set_rotate_label(False)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.set_zlabel(r'$P(\mathbf{x})$', rotation = 90)

ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False

ax.xaxis.pane.set_edgecolor('w')
ax.yaxis.pane.set_edgecolor('w')
ax.zaxis.pane.set_edgecolor('w')

plt.savefig('out/norm_pdf_2d.pgf')
```
