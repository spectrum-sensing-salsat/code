```python
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
dt = 0.00001
t = np.arange(0,1,dt)
f_clean = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)
f_noise = 10*np.random.randn(len(t))
f = f_clean + f_noise

plt.plot(f_clean)
plt.show()
plt.plot(f_noise)
plt.show()
plt.plot(f)
plt.show()
```
